# Visualization of Obesity Levels Data

```python
# Import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read data
df = pd.read_csv("../data/raw/ObesityDataSet_raw_and_data_sinthetic.csv")

print(df.columns)
df.head(5)
```

```python
# TODO: Update df here with cleaned and scaled data set if necessary

# Data sets to plot
df_feature_columns = df.columns.drop('NObeyesdad')
obesity_values = ["Insufficient_Weight", "Normal_Weight", "Overweight_Level_I", "Overweight_Level_II", "Obesity_Type_I", "Obesity_Type_II", "Obesity_Type_III"]
```

```python
# Categorical Plots
import seaborn as sns

df_categorical_columns = ['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS']

def plot_categorical_features(target, feature, data, order):
    plt.figure(figsize = (13, 5))
    sns.countplot(x=target, hue=feature, data=data, order=order)
    plt.show()

for feature in df_categorical_columns:
    plot_categorical_features('NObeyesdad', feature, df, obesity_values)
```

```python
# Numerical Plots
df_numerical_columns = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

def plot_numerical_features(scatter_x, scatter_y):
    plt.figure(figsize = (13, 5))
    plt.scatter(df[scatter_x], df[scatter_y])
    plt.xlabel(scatter_x)
    plt.ylabel(scatter_y)
    plt.tight_layout()
    plt.show()

for num_feature in df_numerical_columns:
    plot_numerical_features('NObeyesdad', num_feature)

```
